```python
def cleanlyrics(dataframe):
    """
    Method to remove special characters in the lyrics column

    input:
    dataframe - Data containing artists and Song Lyrics

    output:
    dataframe - Cleaned Lyrics
    """
    dataframe['Lyrics'] = dataframe['Lyrics'].str.lower()
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"\d", "", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"intro|instrumental|chorus|verse|bridge|outro","", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"\[","", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"\]","", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"  "," ", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"\t","", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(r"[^\w\d'\s]+","", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace(":", "", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace("\n \n \n", "\n", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.replace("\n", " \n ", regex=True)
    dataframe['Lyrics'] = dataframe['Lyrics'].str.strip()
```

```python
# Function to analyze sentiment using the updated ChatGPT API
def get_response(example):
  response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are a sentiment analysis assistant, Provide a sentiment evaluation as 'positive', or 'negative' based on the text. Just reply with Positive or Negative."},
        {"role": "user", "content": example}
    ],
    temperature=1,
    max_tokens=256,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
  )
  return response.choices[0].message.content

def apply_with_progress(row):
    return get_response(row)

# Using tqdm to apply the function and show progress bar
tqdm.pandas(desc="Analyzing Sentiments")
df['Sentiment'] = df['Lyrics'].progress_apply(apply_with_progress)

# Save the results back to a CSV file
df.to_csv('sentiment_output.csv', index=False)

print(df)
```

```python
# Standardize the Sentiment labels
df['Sentiment'] = df['Sentiment'].str.strip()  # Remove any leading/trailing whitespace
df['Sentiment'] = df['Sentiment'].replace(['Negative.', 'Negative'], 'Negative')
df['Sentiment'] = df['Sentiment'].replace(['Positive.', 'Positive ', 'Positive'], 'Positive')

# Drop rows with 'Neutral' sentiment
df = df[df['Sentiment'] != 'Neutral']

# Now df should have only 'Negative' and 'Positive' in the Sentiment column, without any 'Neutral'
df.groupby(df.Sentiment).nunique()
```

```python
senti_sorted = sortedsent = df.groupby(df.Sentiment)
```

```python
#Sorting based on sentiment column value
senti_sortedneg = senti_sorted.get_group('Negative')
senti_sortedpos = senti_sorted.get_group('Positive')
```
